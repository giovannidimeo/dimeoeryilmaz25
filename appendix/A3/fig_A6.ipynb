```python
import pandas as pd
import camelot
```

```python
df19 = pd.DataFrame(columns = ['icd', 
                          'cases_w', 
                          'days_w', 
                          'spell_w', 
                          'cases_m', 
                          'days_m', 
                          'spell_m', 
                          'cases_tot', 
                          'days_tot', 
                          'spell_tot'
                           #,'year'
                              ])
```

```python
tables = camelot.read_pdf(r'***\2019.pdf',
                         pages = "3-26", flavor = 'stream', edge_tol=500)
```

```python
for x in range(29):
    df1 = tables[x].df
    df1 = df1.drop([0,1,2])
    df1 = df1.set_axis(['icd', 
                      'cases_w', 
                      'days_w', 
                      'spell_w', 
                      'cases_m', 
                      'days_m', 
                      'spell_m', 
                      'cases_tot', 
                      'days_tot', 
                      'spell_tot'], axis=1)
    #df1['year'] = 2019
    df19 = df19.append(df1)
    
df19 = df19[:-1]
df19
```

```python
def convert_to_float(value):
    try:
        return float(value)
    except ValueError:
        return None  # Or any other value you want to use for empty cells

for column in df19.columns[1:10]:
    df19[column] = df19[column].apply(str)
    df19[column] = df19[column].str.replace('.','',  regex=False)
    #df19[column] = df19[column].str.strip()
    df19[column] = [convert_to_float(value) for value in df19[column]]
    
df19
```

```python
icd = pd.read_excel(r'***\classification_icd.xlsx',
                    'SA ICD-10 MIT 2021' ) 
icd = pd.DataFrame(icd)
icd = icd[['Chapter_Desc', 'Group_Code', 'Group_Desc', 'ICD10_3_Code', 'ICD10_3_Code_Desc']]

icd.drop_duplicates(subset="ICD10_3_Code", keep="first", inplace=True)
icd
```

```python
df19 = icd.merge(df19, left_on='ICD10_3_Code', right_on='icd',  how='right')
df19
```
